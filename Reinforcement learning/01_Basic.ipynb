### 강화학습이란 ? 
- Agnet가 환경과의 상호작용을 통해 목표를 달성하는 방법을 배우는 문제

- 궁극적인 목표 : Return을 최대화하기 위해 Policy 값을 찾는것 

- 강화학습은 지도학습으로도 가능하지만, 제한되는것이 많다.
    - 머신러닝(기계학습)의 한 방식
    - 지도학습이라고하면 ? state (x) , action (y)

- 강화학습의 용어
    - Agent : 학습의 주체
    - Action : 행동
    - State : 상태 (현재 상황을 기술하는 정보, position)
    - Reword : 보상 (특정 행동에 대해 하나의 숫자로 표현되는 평가 지표)
    - Environment : 환경
    - Policy Function : 방침 함수 (특정 state에서 에이전트가 행동하는 방식 결정)
    - Value Function : 가치 함수 
    - Terminal state : 가장 마지막 상태
    - Return : Action에 대한 (reword * 가중치)를 모두 합친 것

### 특징 
1. 특정 행동에 대한 좋고 나쁨을 평가하는 보상이 주어진다. 
    - 정답은 모르지만 행동에 대한 보상 / 다음의 상태가 주어진다.
    - 보상을 통해 행동이 평가가 된다.
2. 현재의 의사결정이 미래에 영향을 미친다.

3. 문제의 구조를 사전에 알 수 없다.
    - Agent는 수 많은 상호작용을 통해 환경을 예측해볼 수 있다.

- discound Factor
    - r = 0.5, 0.9, 0.99, 0.999
    - 1에 가까운 discound Factor를 사용한다.
    - 0.5가 성능이 더 좋게 나온다.
- Return 값이 클수록 좋다. (최적의 경로)
    - Action이 다를수록 Return 값이 다르다.
- Policy Function
    - 지금상태에서 어떤 Action을 취해야하는지를 알려준다. 
    - State의 함수이다.
