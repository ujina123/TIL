### Linear equation (선형방정식)
- 우리가 흔히 알고 있는 일차 방정식

### determinant (행렬식)
- 행과 열이 있는 매트릭스 형태
- 기하학적으로 넓이(면적)를 구할 수 있다. : 공간에 뿌려보고 면적을 구할 수 있다. 
- 면적이 0일 경우에는 벡터의 방향이 반대일수도 있다.

### System of equations (연립방정식)
- complete(Non-singular, 유용한 의미가 많다), redundant(singular) , contradictory(singular)
- cloumn의 특징(물리, 화학, 국어)에 따라 singular일수도 있다. 
- feature engineering : cloumn 기준으로 데이터를 봐야할 필요가 있다.
