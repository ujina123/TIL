```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = [
    #- 영어,  화학
    [   8 ,  6 ], #- 철수
    [   5 ,  9 ]  #- 영희
]
A = np.asarray(A)
A
```

```python
fig, ax = plt.subplots(1, facecolor="white", figsize=(5,5))
ax.axvline(0)
ax.axhline(0)
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
ax.grid("minor")

#-- subject plot
ax.plot([0,8], [0,6],"ro-") #-- 철수
ax.plot([0,5], [0,9],"bo-") #-- 영희
ax.set_xlabel("English")
ax.set_ylabel("Chmistry")
```

### x,y축이 사람이 될 수 있을까?

```python
fig, ax = plt.subplots(1, facecolor="white", figsize=(5,5))
ax.axvline(0)
ax.axhline(0)
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
ax.grid("minor")

#-- variable plot
#-- 영어점수와 화학점수에 대한 경향성을 볼 수 있다.
ax.plot([0,8],[0,5],"mo-") #-- 영어 점수
ax.plot([0,6],[0,9],"go-") #-- 화학 점수

ax.set_xlabel("young")
ax.set_ylabel("chul")
```

### 매트릭스 연산

```python
A
```

```python
A + 2
```

```python
A * A #-- 수학에서의 2x2매트리스 연산은 아님
```
